# Creating a Phone number column

```python
df['phone_number'] = df['notes'].str.findall('\d{3}[-\.\s]\d{3}[-\.\s]\d{4}|\(\d{3}\)\s*\d{3}[-\.\s]\d{4}|\d{3}[-\.\s]\d{4}')
df['phone_number'] = df['phone_number'].str[0]
df['phone_number'] = df['phone_number'].fillna("None")
df
```

# Creating a Start Date Column

```python
pattern = r'(\d+[-\/\s]?\d{2}[-\/\s]?\d+)|((\d+)?[-\/\s]?(January?|February?|March?|April?|May?|June?|July?|August?|September?|October?|November?|December?)[-\/\s]?\d+)'
```

```python

df['start_date'] = None
inote = df.columns.get_loc('notes')
idate = df.columns.get_loc('start_date')
```

```python
for row in range(0 , 2000):
    start = re.search(pattern,df.iat[row ,inote]).group()
    df.iat[row ,idate] = start
```

#  Creating a End Date Column

```python
pattern2 = r'(end[\s]?\d+[-\/\s]?\d{2}[-\/\s]?\d+)|(end[\s]?(\d+)?[-\/\s]?(January?|February?|March?|April?|May?|June?|July?|August?|September?|October?|November?|December?)[-\/\s]?\d+)'
```

```python
df['End_date'] = None
inote = df.columns.get_loc('notes')
idate_end = df.columns.get_loc('End_date')
```

```python
for row in range(0 , 2000):
    e = re.search(pattern2,df.iat[row ,inote]).group()
    df.iat[row ,idate_end] = e
```
