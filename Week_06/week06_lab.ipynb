## Download & Extract Data

Programmatically download the zip file & extract

```python
file_url = 'https://github.com/msaricaumbc/DS_data/blob/master/lab_data.zip?raw=true'
```

```python
from urllib import request

def download_file(file_name, url):
    res = request.urlopen(url)
    with open(file_name,'wb') as file:
        file.write(res.read())
```

```python
download_file('Lab06.zip',file_url)
```

## Reading data from a csv file

For the file `PoliceShootingsUS.csv`, please answer the following questions:

* How many offenders are older than 55? 

```python
import pandas as pd 
a = pd.read_csv("PoliceShootingsUS.csv",encoding='cp1252')
a
```

```python
c = 0
for x in a.age:
    if x > 55:
        c = c + 1
print("No of offenders are older than 55 is" , c)
```

* What's the age distribution?

```python
## Your code is here
mean = a.age.mean()
mini = a.age.min()
maxi = a.age.max()
print("Avg age of the offender is" , mean)
print("The Youngest offender Age is",mini)
print("The Eldest offender Age is" , maxi)
```

## Dirty data

Read `volunteers.txt` file and perform the following action:

* Create a dataframe with columns: Name, Lastname, State

```python
with open("volunteers.txt","r") as file:
    v = file.readlines()
v
```

```python
Name = [line.split("[")[1] for line in v]
Name = [line.split(",")[0] for line in Name]
Name
```

```python
Lastname = [line.split(",")[1] for line in v]
Lastname = [line.split("]")[0] for line in Lastname]
Lastname
```

```python
state = [line.split("(")[1] for line in v]
state = [line.split(")")[0] for line in state]
state
```

```python
with open("new_volunteer.txt","w") as nf:
    nf.write("Name,Lastname,state\n")
    for i in range(len(Name)):
        nf.write(Name[i])
        nf.write(",")
        nf.write(Lastname[i])
        nf.write(",")
        nf.write(state[i])
        if i != len((Name)) - 1:
            nf.write("\n")
nf
```
