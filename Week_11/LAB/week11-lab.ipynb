```python
html_doc = response.text
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
t = soup.find_all('div', {"class":"col-md-4 country"})
t
```

```python
import pandas as pd
df = pd.DataFrame()
df
d = {"Country":[],"Capital":[],"Population":[],"Area":[]}
```

```python
for i in range(len(t)):
    country_block = t[i]
    d['Country'].append(country_block.find("h3",{'class':'country-name'}).text)
    d['Capital'].append(country_block.find("span",{'class':'country-capital'}).text)
    d['Population'].append(country_block.find("span",{'class':'country-population'}).text)
    d['Area'].append(country_block.find("span",{'class':'country-area'}).text)
```

```python
df = pd.DataFrame(data = d)
```
