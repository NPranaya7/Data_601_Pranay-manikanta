```python
Table_block = t[5]
```

```python
Table_block
```

```python
d = {"Country":[]}
```

```python
for i in Table_block.find_all('tr'):
      d['Country'].append(i.text.strip())
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
df = df['Country'].str.split('\n', expand=True)
```

```python
df
```

```python
df.columns = df.iloc[0] 
```

```python
df = df.drop([0])
```

```python
df
```

```python
df.info()
```

```python
df['2020'] = df['2020'].replace(',', '',regex=True)
df['2021'] = df['2021'].replace(',', '',regex=True)
df['2022'] = df['2022'].replace(',', '',regex=True)
df['2023'] = df['2023'].replace(',', '',regex=True)
df['2024'] = df['2024'].replace(',', '',regex=True)
df['2025'] = df['2025'].replace(',', '',regex=True)
df['2026'] = df['2026'].replace(',', '',regex=True)
df['2027'] = df['2027'].replace(',', '',regex=True)
```

```python
df = df.fillna(0)
```

```python
df
```

```python
df['2020'] = df['2020'].astype(int)
df['2021'] = df['2021'].astype(int)
df['2022'] = df['2022'].astype(int)
df['2023'] = df['2023'].astype(int)
df['2024'] = df['2024'].astype(int)
df['2025'] = df['2025'].astype(int)
df['2026'] = df['2026'].astype(int)
df['2027'] = df['2027'].astype(int)
```

```python
df
```

```python
df['Average Score'] = ((df['2027'] - df['2020'])/df['2020'])*(1/7)*100
```

```python
df
```

```python
df = df.sort_values(by=['Average Score'], ascending=False)
```

```python
df
```

```python
df['2020'] = df['2020'].replace(0, 'None',regex=True)
df['2021'] = df['2021'].replace(0, 'None',regex=True)
df['2022'] = df['2022'].replace(0, 'None',regex=True)
df['2023'] = df['2023'].replace(0, 'None',regex=True)
df['2024'] = df['2024'].replace(0, 'None',regex=True)
df['2025'] = df['2025'].replace(0, 'None',regex=True)
df['2026'] = df['2026'].replace(0, 'None',regex=True)
df['2027'] = df['2027'].replace(0, 'None',regex=True)
```
