## 1. When do the first and last flights leave each day?

```python
g_flights = flights.groupby(['day','month'])
```

```python
first_fight = g_flights.dep_time.min()
Last_fight = g_flights.dep_time.max()
```

```python
print(first_fight,Last_fight)
```

## 2. When there is a missing value in dep_time then this is an indication of canceled flight. Find the number of cancelled flights for each (carrier, origin) combination.


```python
flights['cancel_flight'] = flights.dep_time.isnull()
flight = flights.groupby(['carrier','origin']).sum().astype(int)
print(flight['cancel_flight'])
```

## 3. Which carrier has the worst dep_delays?

```python
g_flights2 = flights.groupby(['carrier'])
worst_flight = flights.groupby(['carrier'])['dep_delay'].mean()
worst_flight
```

```python
worst_flight.idxmax()
```

## 4. Which plane (tailnum) has the worst on-time record?

```python
g_flights3 = flights.groupby('tailnum')['arr_delay'].mean()
g_flights3
```

```python
g_flights3.idxmax()
```

## 5. For each plane, count the number of flights before the first delay of greater than 1 hour.

```python
#flights['cancel_flight'] = flights.dep_time.isnull()

flights['delay'] = flights['dep_delay'] + flights['arr_delay']
print(flights)
```

```python
flights.info()
```

```python
cnt_val,flight_hmap=0,{}
for idx,sort_flight in flights.sort_values(by=['tailnum','delay']).iterrows():
    #print(sort_flight['t'])
    #print(sort_flight['tailnum'])
    dep_delay_s = sort_flight['delay']
    tail_num = sort_flight['tailnum']
    if tail_num not in flight_hmap:
        if dep_delay_s > 60 :
            if tail_num not in flight_hmap:
                flight_hmap[tail_num] = cnt_val
            cnt_val=0
        else:
            cnt_val+=1
print(flight_hmap)
```
