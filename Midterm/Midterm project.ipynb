## **Visulization**

```python
fig, axs = plt.subplots(figsize=(24, 7))
g_df2 =  pd.value_counts(cdf.agencyName).plot(kind='bar', rot=90, ax=axs)
fig.suptitle('No of people in each department',fontsize=25)
plt.xlabel("Departments",fontsize=25)
plt.ylabel("People Count",fontsize=25)
```

```python
fig, axs = plt.subplots(figsize=(24, 7))
g_df2 = cdf.groupby('agencyName')["annualSalary"].mean().plot(kind='bar', rot=90, ax=axs)
fig.suptitle('Avg Annual Salary of each department',fontsize=25)
plt.xlabel("Department",fontsize=25)
plt.ylabel("Salary",fontsize=25)
```

```python
fig, axs = plt.subplots(figsize=(24, 7))
g_df2 = cdf.groupby(['agencyName','agencyID'])["grossPay"].mean().plot(kind='bar', rot=90, ax=axs)
fig.suptitle('Avg Annual Grosspay of each department',fontsize=30)
plt.xlabel("Department",fontsize=25)
plt.ylabel("Grosspay",fontsize=25)
```

```python
fig, axs = plt.subplots(figsize=(24, 7))
g_df2 = cdf.groupby('agencyName')["annualSalary"].plot(kind='hist', rot=90, ax=axs)
fig.suptitle('Avg Annual Salary of each department',fontsize=25)
plt.xlabel("Department",fontsize=25)
plt.ylabel("Salary",fontsize=25)
```

```python
plot = sns.displot(data = cdf, x= 'annualSalary', kind = 'kde', height= 10, aspect = 2).set(title='kernel density estimate of Annual Salary')
sns.set(font_scale = 2)
plt.xlabel("AnnualSalary")
plt.ylabel("Density")
```

```python
plot = sns.displot(data = cdf, x= 'annualSalary', kind = 'ecdf', height= 10, aspect = 2).set(title = 'Empirical distribution function')
#Empirical distribution function
sns.set(font_scale = 2)
plt.xlabel("AnnualSalary")
plt.ylabel("Density")
```

```python
plot = sns.displot(data = cdf, x= 'annualSalary', kind = 'hist', height= 10, aspect = 2).set(title = 'Histogram of Annual Salary')
sns.set(font_scale = 2)
plt.xlabel("AnnualSalary")
plt.ylabel("Count")
```

```python
plot= sns.displot(data=cdf, kind="kde", x="annualSalary", hue = 'agencyName',height= 10, aspect=2, alpha=1).set(title = 'kernel density estimate of Annual Salary by each department')
sns.set(font_scale = 2)
plt.xlabel("AnnualSalary")
plt.ylabel("Density")
plot.ax.grid()
```

```python
plot= sns.displot(data=cdf, kind="kde", x="grossPay", hue = 'agencyName',height= 10, aspect=2, alpha=1).set(title = 'kernel density estimate of GrossPay by each department')
sns.set(font_scale = 2)
plt.xlabel("AnnualSalary")
plt.ylabel("Density")
plot.ax.grid()
```
